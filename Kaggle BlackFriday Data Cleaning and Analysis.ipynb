```python
#We examine if any value in Purchase column in less than 0, which does not make sense
data['Purchase']=data.loc[:,'Purchase']>0
```

```python
data.shape
```

```python
#All purhcases are grouped by User_ID, and ordered by Purchase in descending order
consumers=data.groupby(data['User_ID'])
consumers=consumers.sum().sort_values('Purchase',ascending=False)
consumers.head(10)
#Now we get the top 10 consumers with highest consuming ability
```

```python
#We move on to the next question, what are the characteristics of these consumers?
row1=data[data['User_ID'].isin([1001680])]
row2=data[data['User_ID'].isin([1004277])]
row3=data[data['User_ID'].isin([1001941])]
row4=data[data['User_ID'].isin([1001181])]
row5=data[data['User_ID'].isin([1000889])]
row6=data[data['User_ID'].isin([1003618])]
row7=data[data['User_ID'].isin([1001150])]
row8=data[data['User_ID'].isin([1001015])]
row9=data[data['User_ID'].isin([1002909])]
row10=data[data['User_ID'].isin([1001449])]
new_table=row1.append([row2,row3,row4,row5,row6,row7,row8,row9,row10])
new_table=new_table.reset_index(drop=True)
new_table
```

```python
consumerslist=new_table.drop_duplicates(subset=['User_ID'])
consumerslist=consumerslist.reset_index(drop=True)
consumerslist1=consumerslist.loc[:,['User_ID','Product_ID','Gender','Age','Occupation','City_Category','Marital_Status']]
consumerslist1
```

```python
#We now try to find out the characteristics of these consumers
consumerslist1.apply(pd.value_counts)
```

According to the table above, we could conclude that the main characteristics of these consumers are:
